```python
def chunk_generator(filename, chunk_size):
    with open(filename, "rb") as file:
        while True:
            chunk = [next(file) for _ in range(chunk_size)]
            if not chunk:
                break
            yield chunk
```

```python
def read_figer_data(filename):
    sentences = []
    current_sentence = []

    with open(filename, "rb") as file:
        for line in file:
            line = line.strip()
            if line:
                current_sentence.append(line.split("\t"))
            else:
                if current_sentence:
                    sentences.append(current_sentence)
                    current_sentence = []

    return sentences
```
